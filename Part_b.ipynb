## LR with Gradient Descent

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
# Custom gradient descent optimization for linear regression
def optimize_linear(x, y, learning_rate, iterations, parameters):
    size = x.shape[0]  # number of examples
    weight = parameters["weight"]
    bias = parameters["bias"]

    for i in range(iterations):
        # Predict the output (Linear hypothesis)
        predictions = np.dot(x, weight) + bias
        
        # Calculate the gradients (MSE gradient)
        dW = (1 / size) * np.dot(x.T, (predictions - y))  # Gradient of weights
        db = (1 / size) * np.sum(predictions - y)  # Gradient of bias
        
        # Update parameters using gradient descent
        weight -= learning_rate * dW
        bias -= learning_rate * db

    # Store the updated parameters
    parameters["weight"] = weight
    parameters["bias"] = bias

    return parameters

# 3. Normalize features for better gradient descent performance
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 4. Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=0)

# 5. Initialize parameters
parameters = {
    "weight": np.zeros(X_train.shape[1]),  # Initialize weights with zeros
    "bias": 0  # Initialize bias as zero
}

# 6. Set hyperparameters
learning_rate = 0.1
iterations = 5000

# 7. Run gradient descent optimization
optimized_params = optimize_linear(X_train, y_train, learning_rate, iterations, parameters)

# 8. Make predictions on the test set using optimized parameters
y_pred = np.dot(X_test, optimized_params["weight"]) + optimized_params["bias"]

# 9. Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# 10. Output the evaluation metrics
print("Mean Squared Error (Custom GD):", mse)
print("R-squared (Custom GD):", r2)
```

# Classification

```python
df_full = pd.read_csv("data/zomato_df_final_data.csv")
df['rating_text'] = df_full.rating_text
```

### Binary Encoding

```python
# class_mapping has been applied to create binary_rating
class_mapping = {
    "Poor": 1,
    "Average": 1,
    "Good": 2,
    "Very Good": 2,
    "Excellent": 2
}
df["binary_rating"] = df["rating_text"].map(class_mapping)
df.drop(columns=["rating_text"], inplace=True)

# Convert categorical variables to numerical using one-hot encoding
df = pd.get_dummies(df, drop_first=True)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix


# Define features and target
X = df.drop(columns=['binary_rating'])  # Features
y = df['binary_rating']  # Target

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train the Logistic Regression model
classification_model_logistic_regression_3 = LogisticRegression()
classification_model_logistic_regression_3.fit(X_train, y_train)

# Predictions
y_pred = classification_model_logistic_regression_3.predict(X_test)

# Evaluate
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)  
recall = recall_score(y_test, y_pred)  
f1 = f1_score(y_test, y_pred)  
roc_auc = roc_auc_score(y_test, y_pred)  
conf_matrix = confusion_matrix(y_test, y_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("ROC AUC Score:", roc_auc)
print("Confusion Matrix:\n", conf_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

Out of 945 values 0 ratings, the logistic regression model classified 926 correctly and 19 incorrectly.
On the other hand, from 492 total 1 ratings, the model predicted 38 incorrectly and 454 correctly.
Overall accuracy is 96%, the model predicted 96% of all true positive values suggested by the recall and area under ROC curver is 95%.  
So we can conclude that the model "classification_model_logistic_regression_3" is performing well.
